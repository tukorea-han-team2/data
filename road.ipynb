```python
import requests
import json
import pandas as pd
apiurl = "http://api.vworld.kr/req/data?service=data&request=GetFeature&data=LT_L_SPRD&key=&format=json&size=1000&geomFilter=BOX(124,30,132,40)"
data_list = []
for i in range(1,873):
    print(i)
    params = {'page': i}
    featureCollection={}
    res = json.loads(requests.get(apiurl, params=params).text)
    featureCollection = res["response"]["result"]["featureCollection"]    
    for feature in featureCollection['features']:
        properties = feature['properties']
        rn = properties['rn']
        coordinates = feature['geometry']['coordinates']
    
        for coord_set in coordinates:
            for coord in coord_set:
                longitude, latitude = coord
                data_list.append({'경도': float(longitude), '위도': float(latitude), 'rn': rn})
df = pd.DataFrame(data_list)
print(df)
```

```python
duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
    print("중복된 행이 있습니다:")
    print(duplicate_rows)
    # 중복된 행 삭제
    df.drop_duplicates(inplace=True)
    print("중복된 행을 삭제했습니다.")
else:
    print("중복된 행이 없습니다.")

# 결과 출력
print("최종 데이터프레임:")
print(df)
```

```python
def round_coordinates(data):
    rounded_data = []
    for item in data:
        rounded_item = [round(coord, 7) if isinstance(coord, float) else coord for coord in item]
        rounded_data.append(rounded_item)
    return rounded_data
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = np.array([[row[0], row[1]] for row in rounded_data])  # 경도와 위도를 입력 특성으로 사용
y = np.array([row[2] for row in rounded_data])  # 위치의 이름(주소 또는 도로명)을 출력 타겟으로 사용

# 훈련 및 테스트 세트로 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.00001, random_state=545648556)

# KNN 모델 초기화 및 학습
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)

# 테스트 데이터에 대한 예측
y_pred = model.predict(X_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import numpy as np
def predict_location(model, longitude, latitude):
    # 입력값을 배열 형태로 변환
    input_data = np.array([[longitude, latitude]])
    
    # 모델을 사용하여 예측
    predicted_location = model.predict(input_data)
    
    return predicted_location[0]

# 새로운 경도와 위도 값
new_longitude = 127.0258504
new_latitude = 37.1703277

# 예측된 위치의 이름 출력
predicted_location = predict_location(model, new_longitude, new_latitude)
print("Predicted Location:", predicted_location)
```
