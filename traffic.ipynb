```python
decay_factor = 0.7  # 감쇠 계수 (0과 1 사이의 값)
new_df['acc_year'] = new_df['occrrnc_dt'].astype(str).str[:4].astype(int)
# 2013년부터 2022년까지의 값을 지수적으로 감쇠시킴
current_year = 2022
decay_years = current_year - new_df['acc_year']
new_df['epdo'] = new_df['epdo'] * (decay_factor ** decay_years)
```

```python
def extract_address(address):
    parts = address.split()
    for i, part in enumerate(parts):
        if '-' in part:
            parts[i] = part.split('-')[0]
        if part.endswith(('동', '읍', '리', '가')):
            return address[:address.index(part) + len(part)]
    return address
#읍면리 단위로 주소 분류
```
